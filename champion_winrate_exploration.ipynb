```python
import pandas as pd
import time

GAME_THRESHOLD = 15
dfs = []
for i in range(78, 500, 1):
    print(i)
    URL = "https://na.op.gg/ranking/ladder/page={}".format(i)
    headers = {'User-Agent': 'FooBar-Spider 1.0'}
    response = requests.get(URL, headers=headers, allow_redirects = True)
    response.raise_for_status()
    content = response.text

    soup = BeautifulSoup(response.content, 'html.parser')
    urls = ['https:' + i.find_all('a')[0]['href'] for i in soup.find_all('tr', class_='ranking-table__row')]

    for player_url in urls:
        time.sleep(0.5)
        response = requests.get(player_url, headers=headers, allow_redirects = True)
        response.raise_for_status()
        content = response.text

        soup = BeautifulSoup(response.content, 'html.parser')

        champion_names = [i['title'] for i in soup.find_all('div', class_='MostChampionContent')[0].find_all(class_="ChampionName")]

        champion_winrates = [next(i.descendants).replace('\n', '').replace('\t', '') for i in soup.find_all('div', class_='MostChampionContent')[0].find_all(class_="WinRatio")]
        games_played = [int(next(i.descendants).replace(' Played', '')) for i in soup.find_all('div', class_='MostChampionContent')[0].find_all(class_="Title")]
        
        print(champion_names[0])
        for j in range(len(champion_names)):
            if games_played[j] > GAME_THRESHOLD:
                dfs.append(pd.DataFrame.from_dict({'Champion Name': [champion_names[j]], 'Win Rate': [champion_winrates[j]]}))
                dfs = [pd.concat(dfs)]

                
```

```python
dfs[0].to_csv('champion_winrates2.csv')
```
